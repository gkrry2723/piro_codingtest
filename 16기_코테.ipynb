```python
import random
character = [["김건웅", 65, 16], ["김현주", 82, 20], ["박진혁", 90, 25], ["유송경", 100, 30]]

class Player:
  def __init__(self, name, hp, damage):
    self.name = name
    self.hp = hp
    self.damage = damage
    self.correct_alpha = 0
  

class Game:
  def __init__(self):
    self.round = 0
    self.player = []
    self.word = ["_" for i in range(5)]
    self.answer = ""
    self.player_die = 0
    self.who_last = ""
    self.alpha = [chr(x) for x in range(ord('a'), ord('z'))]

  def start_game(self):
    # 답 생성
    for i in range(len(self.word)):
      self.answer += random.choice(self.alpha)
    print("답: " + self.answer)

    # 캐릭터 생성
    self.player.append(Player("김건웅", 100, 16))
    self.player.append(Player("김현주", 123, 20))
    self.player.append(Player("박진혁", 135, 25))
    self.player.append(Player("유송경", 150, 30))


    print("**************** 게임 접속 ****************")
    print("캐릭터는 다음과 같습니다.")
    for i in range(4):
      print("이름: {0}, HP: {1}, 데미지: {2}".format(self.player[i].name, self.player[i].hp, self.player[i].damage))
    print()
    print()
    # 이건 추가하는건 어떤가 싶어서 넣긴함.
    # print("원하는 Player의 이름을 입력해 주세요: ")
    # while True:
    #   name = input()
    #   if name in ["김건웅", "김현주", "박진혁", "유송경"]:
    #     print(name, "을 선택하셨습니다.")
    #     break
    #   else:
    #     print("없는 player입니다. 위의 4명중에서 선택해주세요.")
  
    print("**************** 게임 시작 ****************")
    print()
    count = 1
    while (not self.player_die) and ("_" in self.word):
      self.start_round(count)
      count += 1
    self.finish_game()

  def start_round(self, count):
    print("===========================")
    print("    ROUND "+ str(count) +" - START")
    print("===========================")
    
    
    # 순서 출력
    if count == 1:
      # 이름순 출력
      self.player.sort(key = lambda x: x.name) 
    else:
      self.player.sort(key = lambda x: x.hp, reverse=True)
    
    print("게임은 {0}, {1}, {2}, {3} 순으로 진행됩니다.".format(self.player[0].name, self.player[1].name, self.player[2].name, self.player[3].name))
    for i in range(4):

      print("— 내 캐릭터 —")
      print("이름: {0} (HP: {1})".format(self.player[i].name, self.player[i].hp))
      choice_alpha = random.choice(self.alpha)
      self.alpha.pop(self.alpha.index(choice_alpha))

      print("선택 알파벳: "+ choice_alpha)
      if choice_alpha in self.answer:
        self.player[i].correct_alpha += 1
        print("— 맞았습니다 ᵔεᵔ  —")
        for j in range(len(self.word)):
          if self.answer[j] == choice_alpha:
            self.word[j] = choice_alpha
        if "_" not in self.word:
          self.who_last = self.player[i].name
          break  
      else:
        print("— 틀렸습니다 (ﾟ⊿ﾟ) —")
        self.player[i].hp -= self.player[i].damage
        print("{0}님은 틀렸기 때문에 HP가 {1} 남았습니다.".format(self.player[i].name, self.player[i].hp))

        # 게임 종료 조건 탐색
        if self.player[i].hp <= 0:
          self.player_die= 1
          self.who_last = self.player[i].name
          break
        if "_" not in self.word:
          self.who_last = self.player[i].name
          break


    print("============================")
    print("      ROUND {} - END".format(count))
    print("===========================")
    print("남은 상태: {}".format("".join(self.word)))
    print()
    print("현재 캐릭터의 상태는 다음과 같습니다.")
    for i in range(4):
      print("이름: {0}, HP: {1}, 맞춘 개수: {2}".format(self.player[i].name, self.player[i].hp, self.player[i].correct_alpha))
    print()
    print()
    print()


  def finish_game(self):
    print("<<<< 게임이 끝났습니다 >>>>")
    if self.player_die == 1:
      print("{0}님이 사망하여 게임이 끝났습니다.".format(self.who_last))
    else:
      print("{0}님이 마지막 글자를 맞췄습니다.".format(self.who_last))
    print("=============================")
    print("     게임 순위 - 생명력")
    print("=============================")
    self.player.sort(key = lambda x: x.hp, reverse=True)
    for i in range(4):
      if self.player[i].hp < 0:
        print("{0}등 {1} (사망자)".format(str(i+1), self.player[i].name))
      else:
        print("{0}등 {1} (HP: {2})".format(str(i+1), self.player[i].name, self.player[i].hp))
    
    print("=============================")
    print("게임 순위 - 알파벳 맞춘 횟수")
    print("=============================")
    self.player.sort(key = lambda x: x.correct_alpha, reverse=True)
    for i in range(4):
      print("{0}등 {1} {2}회".format(str(i+1), self.player[i].name, self.player[i].correct_alpha))
      


```
